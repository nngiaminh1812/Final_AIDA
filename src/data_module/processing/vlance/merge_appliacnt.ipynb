```python
import pandas as pd
import glob
import os
```

```python
# Đường dẫn đến thư mục chứa các file cần hợp nhất
folder_path = r'..\..\..\..\data\raw\applicant'

# Tìm tất cả các file CSV trong thư mục
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Danh sách để lưu dữ liệu từ các file
dataframes = []

# Đọc và hợp nhất từng file
count = 0
for file in all_files:
    print(f"Đang hợp nhất file: {os.path.basename(file)}")  # In tên file đang xử lý
    df = pd.read_csv(file)
    dataframes.append(df)
    count += 1

# Kết hợp tất cả các DataFrames
merged_df = pd.concat(dataframes, ignore_index=True)

# Lưu file hợp nhất thành applicant.csv
output_path = r'..\..\..\..\data\raw\applicant.csv'
merged_df.to_csv(output_path, index=False)

print(f"Tất cả các file đã được hợp nhất thành {output_path}, đã hợp nhất {count} file")
```
