```python
import pandas as pd
import os
import numpy as np
```

```python
# list all files in current directory
files = os.listdir('.')
files
```

```python
# list all files that start with 'job_posts_' and end with '.csv'
job_post_files = [file for file in files if file.startswith('job_posts_') and file.endswith('.csv')]
job_post_files
```

```python
# read all files into a list of dataframes
dataframes = []
for file in job_post_files:
    df = pd.read_csv(file, encoding='utf-8')
    dataframes.append(df)

# concatenate all dataframes into a single dataframe
merged_df = pd.concat(dataframes, ignore_index=True)
```

```python
# write the merged dataframe to a new CSV file
merged_df.to_csv('merged_job_posts.csv', index=False)
```
